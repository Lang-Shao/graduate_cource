```julia
using StatsBase, PyPlot
names = ["Mary","Mel","David","John","Kayley","Anderson"];
randomName() = rand(names)
X = 3:8
N = 10^6
sampleLengths = [length(randomName()) for _ in 1:N]
bar(X,counts(sampleLengths)/N);
xlabel("Name length")
ylabel("Estimated p(k)")
```

```julia
using PyPlot
pDiscrete = [0.25, 0.25, 0.5]

xGridD = 0:2
pContinuous(x) = 3/4*(1 - x^2)

xGridC = -1:0.01:1
pContinuous2(x) = x < 0 ? x+1 : 1-x

figure(figsize=(12.4,4))
subplots_adjust(wspace=0.4)
subplot(131)
#stem(xGridD,pDiscrete,basefmt="none")
bar(xGridD,pDiscrete)
ylim(0,1.1)
xlabel("x")
ylabel("Probability")

subplot(132)
plot(xGridC,pContinuous.(xGridC))
ylim(0,1.1)
xlabel("x")
ylabel("Density")

subplot(133)
plot(xGridC,pContinuous2.(xGridC))
ylim(0,1.1)
xlabel("x")
ylabel("Density");
```

```julia
using QuadGK
sup = (-1,1)
f1(x) = 3/4*(1-x^2)
f2(x) = x < 0 ? x+1 : 1-x
expect(f,support) = quadgk((x) -> x*f(x),support[1],support[2])[1]
expect(f1,sup),expect(f2,sup)
```
