#  Two methods for accomplishing Color shift

```python
from PIL import Image
import matplotlib.pyplot as plt
import scipy.misc
import imageio
import scipy.ndimage as nd

face=scipy.misc.face()
plt.imshow(face)
imageio.imsave( 'data/face.jpg', face )
```

```python
im = Image.fromarray(face)
r,g,b = im.split()
im2 = Image.merge('RGB', (g,b,r) )
plt.imshow(im2)
```

```python
import imageio
import scipy.ndimage as nd
im3 = nd.shift( im, (0,0,1), mode='wrap' )
plt.imshow(im3)
```

```python
im4 = 0.5 *face[:,:,0] + 0.75*face[:,:,1] + 0.25*face[:,:,2]
plt.imshow(im4)
```
