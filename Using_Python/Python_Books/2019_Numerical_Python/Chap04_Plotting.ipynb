```python
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
matplotlib.__version__
%matplotlib inline
import numpy as np
import sympy
```

```python
x = np.linspace(-5,2,100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x + 10
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(x[::4], y1[::4], color="blue", label="y(x)", linestyle='-.',marker='.')
#ax.plot(x, y2, color="red", label="y'(x)")
axes[0,0].plot(x[::30], y2[::30], color="red", label="y'(x)")
axes[0,0].plot(x, y3, color="green", label="y''(x)")
axes[0,0].scatter(x[::2], 2*y1[::2], color="orange", label="y''(x)")
axes[0,0].set_xlabel("T (s)")
axes[0,0].set_ylabel("y (Km)")
axes[0,0].legend()
```
