```python
from scipy import optimize
import numpy as np

def fun7_12(R):
    return R**3-(7+4*np.sqrt(2))*R**2+(3+4*np.sqrt(2))*R-1

sol = optimize.fsolve(fun7_12,10.0)
print(sol)
```

```python
def del_vH(R):
    return (1-1/R)*(2*R/(1+R))**0.5+R**(-0.5)-1

def del_vBP(R):
    return (np.sqrt(2)-1)*(1+R**(-0.5))

def fun7_12(R):
    return del_vH(R)-del_vBP(R)

sol = optimize.fsolve(fun7_12,1.0)
print(sol)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(1,20,100)
plt.plot(x, del_vH(x),label='Hohmann')
plt.plot(x, del_vBP(x),label='Bi-Parabolic')
plt.xlabel('R=r2/r1')
plt.ylabel('delta V/ Vcircle')
plt.axvline(sol[0],linestyle='--',label=f"{sol[0]:0.2f}")
plt.legend()
```
